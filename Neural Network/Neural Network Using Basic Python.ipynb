# Neural Network using simple python and numpy

```python
x=np.array([200,17])
# x.reshape(-1,1)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

### Layer 1 Activation 1_1

```python
w1_1=np.array([1,2])
b1_1=np.array([-1])
m=x.shape[0]
z1_1=np.zeros(m)
z1_1=np.dot(w1_1,x)+b1_1
a1_1=sigmoid(z1_1)
```

```python
print(a1_1)
```

### Layer 1 Activation 1_2

```python
w1_1=np.array([-3,4])
b1_1=np.array([1])
m=x.shape[0]
z1_2=np.zeros(m)
z1_2=np.dot(w1_1,x)+b1_1
a1_2=sigmoid(z1_2)
```

```python
a1_2
```

### Layer 1 Activation 1_3

```python
w1_1=np.array([5,-6])
b1_1=np.array([2])
m=x.shape[0]
z1_3=np.zeros(m)
z1_3=np.dot(w1_1,x)+b1_1
a1_3=sigmoid(z1_3)
```

```python
a1_3
```

```python
# Final Activation:a_1

a_1=np.array([a1_1,a1_2,a1_3])
```

### Layer 2 /Output

```python
w2_1=np.array([-7,8,9])
b2_1=np.array([3])
z2_1=np.dot(w2_1,a_1)+b2_1
a2=sigmoid(z2_1)
```

```python
a2
```

# Neural Network General Implementation

```python
w=np.array([
    [1,-3,5],
    [2,3,-6]])
b=np.array([-1,1,2])
a_in=np.array([-2,4])
```

```python
def dense(a_in,w,b):
    units=w.shape[1]
    a_out=np.zeros(units)
    for j in range(units):
        w_j=w[:,j]
        z=np.dot(w_j,a_in)+b[j]
        a_out[j]=sigmoid(z)
    return a_out
 
def sequential():
    w1 = np.array([[1,-3,5],[2,3,-6]])
    b1 = np.array([-1,1,2])
    w2 = np.array([[4],[-2],[1]])
    b2 = np.array([0])
    a_in=np.array([-2,4])
    
    a1 = dense(a_in, w1, b1)
    a2 = dense(a1, w2, b2)
    f_x = a2
    
    return f_x
```

```python
final=sequential()
final
```
