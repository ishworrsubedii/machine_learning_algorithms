# Ordinary Least Squares (OLS) Linear Regression
The Ordinary Least Squares (OLS) method is used to find the best-fit line for a set of data points by approximating a function and taking its derivative.

## Understanding the Goal

- Ordinary Least Square aims to find the line of best fit for a set of data points.
- This line is represented by the equation `y = w*x + b`, where:
    - `y` is the target variable
    - `x` is the input variable
    - `w` is the weight (slope)
    - `b` is the bias (y-intercept)

## Using Derivatives to Minimize Errors

1. **Define the Cost Function:**
   - The Sum of Squared Errors (SSE) measures the "goodness" of the fit:
      - SSE = (y_actual - y_predicted)^2

2. **Calculate Derivatives:**
   - Find the values of w and b that minimize SSE:
      - SSE/w = 0
      - SSE/b = 0
## Equation for finding the slope and intercept
After derivation we can det this equation





```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train=np.array([2,4,8,9,10,12,13,16])
Y_train=np.array([5,9,17,20,23,26,28,32])
```

```python
def ordinary_least_squares(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sum((x - x_mean) ** 2)
    
    w = numerator / denominator
    b = y_mean - (w * x_mean)
    
    return w, b

```

```python
w,b=ordinary_least_squares(X_train, Y_train)
print(w,b)

def prediction(x_train, w, b):
    predictions = []
    for i in x_train:
        prediction = w * i + b
        predictions.append(prediction)
    return predictions
```

```python
y_pred = prediction(X_train, w, b)

plt.scatter(X_train, Y_train)
plt.plot(X_train, y_pred, color='red')
plt.show()
```
